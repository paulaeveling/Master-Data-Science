```python
# importar dados csv ou planilha
file_path = ""

doc = pd.read_csv(file_path,
                 header=None,
                 names=[
                     'sepal_lenght',
                     'sepal_width',
                     'petal_length',
                     'petal_width',
                     'species'
                 ])

# imprimir 10 primeiras linhas
doc.head(10)

# imprimir 10 ultimas linhas
doc.tail(10)

# descrever documento incluindo todos os dados
doc.describe(include="all")

# descrever documento incluindo somente nao numericos
doc.describe(include="object")

# descrever documento incluindo somente valores numericos
doc.describe()

# descrever documento incluindo somente  numericos
doc.describe(include="numbers")
```

```python
# retornar shape do documento (numero de linhas/colunas)
doc.shape

# acessar numero de colunas ou linhas
doc.column
doc.index

# acessar elemento dentro do documento
doc.sepal_lenght

# transformar em matriz numpy ndarray
doc.as_matrix()

# transformar em np.array de 1 dimensao
type(doc['species'].as_matrix())
```
