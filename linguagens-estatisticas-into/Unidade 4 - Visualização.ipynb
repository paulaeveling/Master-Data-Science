```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats

%matplotlib inline

np.random.seed(123)
```

```python
x = np.abs(np.random.normal(60,500,100))[:,np.newaxis]
x[x<=40] = 40
w = 2000
y = x * (w - np.random.normal(0,500, x.shape[0]))[:,np.newaxis] + 60000

# funcao para desenhar pontos
plt.scatter(x,y)
```

```python
x_1 = np.hstack((np.ones((x.shape[0],1)),x))

# estima inclinacao com metodo dos minimos quadrados
b,w = np.linalg.inv(x_1.T.dot(x_1)).dot(x_1.T).dot(y).ravel()

plt.scatter(x,y)
plt.plot(x, w * x + b,color="green")

# scatter = pontos | plot = linhas
```

# PLOTNINE

```python
from plotnine import *

x = np.abs(np.random.normal(60,500,100))[:, np.newaxis]
x[x <= 40] = 40
w = 2000
y = x * (w - np.random.normal(0, 500, x.shape[0]))[:, np.newaxis] + 6000

df = pd.DataFrame(np.hstack((x,y)), columns=['area','preco'])
df.head()
```

```python
# aes = aesthetic

(ggplot(df, aes(x = 'area', y = 'preco')) +
 geom_point())
```

```python
# se == precisao / margem de erro

(ggplot(df, aes(x = 'area', y = 'preco')) +
 geom_point() +
 geom_smooth(method='lm', se = True))
```
