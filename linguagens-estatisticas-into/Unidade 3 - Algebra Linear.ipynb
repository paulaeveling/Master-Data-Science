```python
import numpy as np

a = np.array([[1.0,2.0],[3.0,4.0]])
print("matriz a:", a)
```

```python
# matriz transposta
print("matriz transposta 1:\n",a.transpose())
print("matriz transposta 2:\n",a.T)
```

```python
# inversao de matriz
print(np.linalg.inv(a))
```

```python
# matriz e identidade - todos numeros sao 0 exceto diagonal principal
print(np.eye(10))
```

```python
# multiplicacao de matriz
a = np.array([[1.0,2.0],[3.0,4.0]])
print("multiplicacao de matriz:\n",np.dot(a, np.linalg.inv(a)))
```

```python
# calculo de autovalores e autovetores
print(np.linalg.eig(a))
```
