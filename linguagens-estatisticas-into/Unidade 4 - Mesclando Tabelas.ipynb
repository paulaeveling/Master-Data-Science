```python
import pandas as pd

left = pd.DataFrame({
    'KEY': ['KO', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

right = pd.DataFrame({
    'KEY': ['KO', 'K1', 'K2', 'K3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

print(left)
print("")
print(right)
```

```python
# mesclar tabela left e right
# lembrar que python eh case sensitive

pd.merge(left, right, on="KEY")
```

```python
left2 = pd.DataFrame({
    'KEY1': ['KO', 'K0', 'K1', 'K2'],
    'KEY2': ['KO', 'K1', 'K0', 'K1'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

right2 = pd.DataFrame({
    'KEY1': ['KO', 'K1', 'K1', 'K2'],
    'KEY2': ['KO', 'K0', 'K0', 'K0'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

print(left2)
print("")
print(right2)
```

```python
# nesse exemplo, mantem-se as duas chaves
# somente valores que match as keys eh que serao mesclados
# inner join
pd.merge(left2, right2, on=["KEY1", 'KEY2'])
```

```python
# left join
# mantem dados da tabela da esquerda
pd.merge(left2, right2, how='left', on=['KEY1', 'KEY2'])
```

```python
# right join
# mantem dados da tabela da direita
pd.merge(left2, right2, how='right', on=['KEY1', 'KEY2'])
```

```python
print(left2)
print("")
print(right2)

# join: junta todos os dados basedao na key
pd.merge(left2, right2, how='outer', on=['KEY1', 'KEY2'])
```

```python
print(left2)
print("")
print(right2)

# adicionando mais um join...
pd.merge(pd.merge(left2, right2, how='outer', on=['KEY1', 'KEY2']), left2, on=['KEY1', 'KEY2'])
```
